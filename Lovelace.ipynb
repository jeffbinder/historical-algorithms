```python
def check(m, n, d, mp, np, dp, x, y):
    print('{0}x + {1}y = {2}'.format(m, n, d))
    print('{0}({3}) + {1}({4}) = {2}'.format(m, n, m*x + n*y, x, y))
    print('{0}x + {1}y = {2}'.format(mp, np, dp))
    print('{0}({3}) + {1}({4}) = {2}'.format(mp, np, mp*x + np*y, x, y))
```

```python
def solve(m, n, d, mp, np, dp):
    V = [0] * 17
    V[0] = m
    V[1] = n
    V[2] = d
    V[3] = mp
    V[4] = np
    V[5] = dp
    V[6] = V[0] * V[4]
    print(V)
    V[7] = V[1] * V[3]
    print(V)
    V[8] = V[2] * V[4]
    print(V)
    V[9] = V[5] * V[1]
    V[1] = 0
    print(V)
    V[10] = V[0] * V[5]
    V[0] = 0
    V[5] = 0
    print(V)
    V[11] = V[2] * V[3]
    V[2] = 0
    V[3] = 0
    print(V)
    V[12] = V[6] - V[7]
    V[6] = 0
    V[7] = 0
    print(V)
    V[13] = V[8] - V[9]
    V[8] = 0
    V[9] = 0
    print(V)
    V[14] = V[10] - V[11]
    V[10] = 0
    V[11] = 0
    print(V)
    V[15] = V[13] / V[12]
    V[13] = 0
    print(V)
    V[16] = V[14] / V[12]
    V[14] = 0
    V[12] = 0
    print(V)
    return V[15], V[16]
```

```python
m = 2
n = 6
d = 3
mp = 7
np = 5
dp = 7
x, y = solve(m, n, d, mp, np, dp)
check(m, n, d, mp, np, dp, x, y)
```
