```python
from math import *
```

```python
def a_1(u, r, t):
    return u / r**t
```

```python
a_1(364.927, 1.06, 9)
```

```python
def a_2(u, r, t):
    log_r = log10(r)
    print("r = {0} \t\t its Logarithm {1}".format(r, log_r))
    print("t = \t\t\t {0}".format(t))
    print("\t\t\t ------")
    log_rt = log_r * t
    print("r(t) = a Number \t its Logarithm {0}".format(log_rt))
    log_u = log10(u)
    print("u = {0} \t\t its Logarithm {1}".format(u, log_u))
    print("\t\t\t ------")
    log_a = log_u - log_rt
    a = 10 ** log_a
    print("a = {0} \t its Logarithm {1}".format(a, log_a))
    return a
```

```python
a_2(364.927, 1.06, 9)
```
